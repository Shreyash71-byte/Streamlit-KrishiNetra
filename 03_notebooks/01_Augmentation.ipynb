```python
import os                # Used for: folders access, file paths, directory operations
import tensorflow as tf           # deep learning framework, CNN building, training & prediction.
from tensorflow.keras import layers, models      # Used for: Conv2D, Dense and Sequential CNN architecture building.

from tensorflow.keras.preprocessing.image import ImageDataGenerator      # This class handles: image loading, preprocessing, augmentation & batching Automatically.
```

```python
# Instead of manually: opening image, resize image, normalize image, rotate image ImageDataGenerator automates all of these automatically
```

```python
# SECTION 1 — TRAIN DATA AUGMENTATION : “image transformation machine” ----->  image processing pipeline ban rahi hai.

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True 
    )
```

```python
# train_datagen   = ImageDataGenerator(...): 
# This creates an instance of the ImageDataGenerator class. It defines the rules for how images will be modified as they are loaded, but it does not modify them yet. 

# rescale = 1./255 : 
# This is a normalization step. Input image pixels typically have values from 0 to 255 (RGB). This multiplies every pixel by 1/255 ---> approx 0.0039, changing the pixel value range from [0, 255] to [0, 1]. 
# Since Neural networks large values pe unstable ho sakte hain. This makes the input data more suitable for neural networks, allowing them to converge faster. 

# rotation_range =20: 
# This instructs the generator to randomly rotate the image between (-20) and (+20) degrees. This makes the model robust to objects that are slightly tilted.

# zoom_range =0.2: 
# This instructs generator to randomly zoom in or out of the image by up to (20%) i.e., (0.2). If specified as a float, it means a zoom range of ([1-0.2, 1+0.2]), i.e., between (0.8) and (1.2).

# horizontal_flip=True: 
# This randomly flips the image horizontally (left-to-right). This is useful when the object orientation does not change its class (e.g., a car facing left vs. a car facing right are both cars)
```

Summary of What Happens:
During training, every time an image is loaded, it will be:
- Normalized (divided by 255).
- Randomly rotated within (p/m 20) degrees.
- Randomly zoomed in/out.
- Randomly flipped horizontally.
- Augmentation helps prevent overfitting because model sees: slightly different versions each epoch.
- This means the network rarely sees the exact same image twice, forcing it to learn more robust features.

```python
# SECTION 2 : VALIDATION DATA
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Validation data should NOT be augmented.
# Validation should represent: real unseen. If augmented: evaluation becomes unrealistic.
# So only normalization applied.
```

```python
# SECTION 3: TRAIN GENERATOR
train_generator = train_datagen.flow_from_directory(
    '02_dataset/01_train',
    target_size=(150, 150),
    batch_size=4,
    class_mode='categorical'
)
```

```python
# train_generator = train_datagen.flow_from_directory(...): 
# This creates a Python iterator that streams images directly from a local directory. It uses the augmentation rules (like rotation and zoom) we defined previously in train_datagen.

# '02_dataset/01_train' :
# This is the path to your training folder. Keras expects this directory to contain subfolders (e.g., dog/, cat/), where each subfolder name acts as the label for the images inside it.

# target_size=(150, 150):
# This automatically resizes every image to (150 times 150) pixels as it is loaded. This ensures all input data has uniform dimensions, which is a strict requirement for neural networks.
# Smaller images: faster training ------> but less detailed.
# Larger images : better detail --------> slower training.

# batch_size=4:
# This groups your images into batches of 4. The model will look at 4 images at a time, calculate the loss, and update its weights before moving to the next 4 images.
# Batch size affects : memory usage, speed & learning stability.

# class_mode='categorical':
# This tells the generator you are doing multi-class classification (3 or more categories). It automatically converts your folder-name labels into one-hot encoded vectors (e.g., [1, 0, 0], [0, 1, 0]).
```

- Summary of What Happens :
- When our model runs, this generator will continuously scan '02_dataset/01_train', grab 4 images at a time, resize them to (150 X 150), apply our augmentations.
- Convert their labels to one-hot vectors, and feed them into the network.

```python
# SECTION 4: VALIDATION GENERATOR
val_generator = val_datagen.flow_from_directory(
    '02_dataset/01_train',
    target_size=(150, 150),
    batch_size=4,
    class_mode='categorical'
)

# Same process but no augmentation, only normalization.
```
